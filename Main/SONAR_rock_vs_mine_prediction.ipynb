Importing the libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Converting to data frames

```python
data=pd.read_csv("/content/drive/MyDrive/Sonar Rock vs Mine Prediction/Copy of sonar data.csv", header=None)
```

```python
x=data.drop(columns=60,axis=1)
y=data[60]
```

```python
print(x.shape)
```

```python
print(y.shape)
```

Training and Test


```python
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)
```

```python
print(y_train)
```

Model Training

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
data_pre=model.predict(x_train)
acc=accuracy_score(data_pre, y_train)
```

```python
print(acc)
```

```python
data_predict=model.predict(x_test)
accuraccy_test=accuracy_score(data_predict, y_test)
print(accuraccy_test)
```
